#4. Sharpe Ratio

##**tells you how much return you are getting for each unit of risk you are taking.**
- A higher Sharpe ratio means that you are getting more return for your risk.
**A Sharpe ratio of 1 is considered average.**
**A Sharpe ratio of 2 or higher is considered good.**
**A Sharpe ratio of 3 or higher is considered excellent.**
- The risk-free rate is the return of an investment that is considered to be free of risk.
- The standard deviation is a measure of how volatile an investment is.

##Cons:
- It does not take into account the risk of loss (of entire trading account?).
- It is not always easy to calculate the risk-free rate.
- It can be misleading for investments with very different volatilities.

- Despite its limitations, the Sharpe ratio is a useful tool for comparing the risk-adjusted performance of different investments.
